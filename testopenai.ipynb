```python
## Chat Completion API and Completion API

## Completion API

response =openai.Completion.create(
    model='gpt-3.5-turbo-instruct',
    prompt='Explaining how to make a sandwich.',
    max_tokens=25
)
response 
```

```python
response =openai.Completion.create(
    model='gpt-3.5-turbo-instruct',
    prompt='Explain machine learning to a 5 year old.',
    max_tokens=25,
    temperature=0.6,
    n=5
)
response 
```

```python
print(response.choices[0]['text'])
```

```python
response=openai.Completion.create(
    model='gpt-3.5-turbo-instruct',
    prompt="""Act as an AI model to detect positive or negative tweets:
    "This person is bad and he is biased towards negative sentiments"
    """,
    max_tokens=25,
    temperature=1,
    n=1
)
print(response)
```

```python
response=openai.Completion.create(
    model='gpt-3.5-turbo-instruct',
    prompt="""Act as an AI model to translate text to French:
    "Hello How are you?"
    """,
    max_tokens=25,
    temperature=1,
    n=1
)
print(response)
```

```python
## Chat Completion API

openai.ChatCompletion.create(
    model='gpt-3.5-turbo',
    messages=[{'role':'system','content':'Act as an AI assistant'},
              {'role':'user','content':'Hello how are you?'},
              {'role':'user','content':'My age is 20 and I am a Data Scientist'},
              {'role':'user','content':'What is my age'}

    ]
)
```
